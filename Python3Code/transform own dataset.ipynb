```python
from pathlib import Path
import pandas as pd
```

```python
acc_df = pd.read_csv("./datasets/crowdsignals/csv-participant-one/accelerometer_phone.csv")
own_df = pd.read_csv("./datasets/own/nothing/Accelerometer.csv")
own_meta_df = pd.read_csv("./datasets/own/nothing/meta/time.csv")
```

```python
acc_df
```

```python
start_time = own_meta_df[own_meta_df.event == "START"]["system time"][0]
```

```python
own_df = own_df.rename(columns={
    "Time (s)": "timestamps",
    "Acceleration x (m/s^2)": "x",
    "Acceleration y (m/s^2)": "y",
    "Acceleration z (m/s^2)": "z",
})
own_df
```

```python
own_df["timestamps"] += start_time
own_df["timestamps"] = own_df["timestamps"] * 10**9
own_df["timestamps"] = own_df["timestamps"].astype(int)
own_df
```

```python
# make sure the epochs are the same format
assert len(str(acc_df["timestamps"][0])) == len(str(own_df["timestamps"][0]))
```
